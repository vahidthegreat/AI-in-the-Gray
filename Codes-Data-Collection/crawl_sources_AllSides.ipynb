```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import NoSuchElementException
from tqdm.notebook import tqdm
from selenium.webdriver.common.keys import Keys
import pandas as pd

import traceback
import requests
import csv
import urllib.parse
import time
import os
from bs4 import BeautifulSoup

from selenium.webdriver import FirefoxOptions

firefox_user_path = "C:\\Users\\username\\AppData\\Roaming\\Mozilla\\Firefox\\Profiles\\userprofile"
options=webdriver.FirefoxOptions()
options.profile = firefox_user_path
# options.add_argument("--headless")
options.set_preference('profile', firefox_user_path)
driver = webdriver.Firefox(options=options)

driver.get(base_url)
# driver.implicitly_wait(20)
time.sleep(3)
driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/button[2]').click()
time.sleep(3)
# driver.imfind_elementplicitly_wait(20)
driver.(By.XPATH,'/html/body/div[8]/div/button').click()
time.sleep(3)

meta_data_df=pd.DataFrame()

# body.send_keys(Keys.ARROW_DOWN)
time.sleep(10)
# driver.find_element(By.XPATH,"/html/body/div[5]/div[1]/div/div[2]/div/div/div/div[3]/small/div[1]/a/p[1]").click()
time.sleep(1)
body = driver.find_element(By.TAG_NAME,'body')
while True:
#     while True:
#         try:
#             driver.find_element(By.XPATH,"/html/body/div[4]/div[1]/div/div[2]/div/div/div/div[3]/small/div[1]/a").click()            
#             time.sleep(3)
#             break
#         except:
#     new_height = driver.execute_script("return document.body.scrollHeight / 2")
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(5)
    try:
        driver.find_element(By.XPATH,"/html/body/div[4]/div[1]/div/div[2]/div/div/div/div[3]/small/div[1]/a/p[1]").click()
    except:
        None
    posts = driver.find_elements(By.XPATH,"/html/body/div[5]/div[1]/div/div[2]/div/div/div/div[2]/table/tbody/tr")

        #     new_posts = soup.find_all("li", class_="card-grid__card card-grid__card--square")
    for post_i in posts[len(meta_data_df):]:
        post_i_soup = BeautifulSoup(post_i.get_attribute('outerHTML'), "html.parser")
        title_x = post_i_soup.find("td", class_="views-field views-field-title source-title").find('a').text
        url_x = post_i_soup.find("td", class_="views-field views-field-title source-title").find('a').get("href")
        leaning = post_i_soup.find("td", class_="views-field views-field-field-bias-image").find('a').get("href").split('/')[-1]
        agree = post_i.find_element(By.XPATH,".//td[4]/div/div[1]/div[2]/span[1]").text
        disagree = post_i.find_element(By.XPATH,".//td[4]/div/div[1]/div[2]/span[2]").text
        try:
            community_confidence = post_i.find_element(By.XPATH,".//td[4]/div/div[2]/span").text
        except:
            community_confidence = ''
        meta_data_df = meta_data_df.append({'source_name':title_x,'url':url_x,'leaning':leaning,'agree':agree,'disagree':disagree,'community_confidence':community_confidence}, ignore_index=True)
    print(str(len(meta_data_df))+' -> '+str(len(meta_data_df.drop_duplicates())))
```

```python
import re

def domain_extractor(url_x):
    url_x = re.sub('https://','',url_x)
    url_x = re.sub('www\.','',url_x)
    url_x = re.sub('/$','',url_x)
    return url_x

def get_missing_websites(url_x,info_dict):
    
    if 'Website' not in info_dict:

        driver.get(base_url+url_x)
        print('Trying for:')
        print(base_url+url_x)
        try:
    #         time.sleep(3)
    #         driver.find_element(By.XPATH,'/html/body/div[1]/div/div/div/div[2]/div/button[2]').click()
            time.sleep(1)
            driver.find_element(By.XPATH,'/html/body/div[8]/div/button').click()
        except:
            None
        website = driver.find_element(By.CLASS_NAME,"external-link").get_attribute('href')
        print(website)
        website=domain_extractor(website)
        print(website)
        info_dict['Website'] = website

        print(info_dict)
        print('\n')
    return info_dict


meta_data_df['other_info'] = meta_data_df.progress_apply(lambda x: get_missing_websites(x['url'],x['other_info']) ,axis=1)
```
