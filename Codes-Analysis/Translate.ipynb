# Translate_Meta_Data

```python
def translate_to_english(text):
  translator = Translator()
  translation = translator.translate(text, dest="en")
  print(translation.text)
  return pd.Series([translation.src,translation.text])  

def translate_tags(tags_list_x, language_x):
  if language_x == 'en':
    return tags_list_x.lower()
  else:
    tags_list_x=tags_list_x.split(',')
    tags_list_x_translated = []
    for tag_x in tags_list_x:
      translator = Translator()     
      translation = translator.translate(tag_x, dest="en")
      time.sleep(0.1)
      tags_list_x_translated.append(translation.text.strip().lower())
    tags_list_x_translated = ','.join(tags_list_x_translated)
    print(tags_list_x_translated)
    return tags_list_x_translated
```
